```python
# list_comprehensions_1

X = int(input())
Y = int(input())
Z = int(input())
N = int(input())

RESULT = []

for a in range(0, X + 1):
    for b in range(0, Y + 1):
        for c in range(0, Z + 1):
            if (a + b + c) != N:
                RESULT.append([a, b, c])

print(RESULT)
```

```python
# nested_lists_1

if __name__ == '__main__':
    
    students = []
    
    for _ in range(int(input())):
        name = input()
        score = float(input())
        students.append([name, score])
    mn = min(students, key=lambda x: x[1])
    nonlowest = sorted([student for student in students if student[1] > mn[1]], key= lambda x: x[1])
    seconds = sorted([student[0] for student in nonlowest if student[1] == nonlowest[0][1]])
    for student in seconds:
        print(student)
```

```python
# nested_lists_2

STUDENTS = []
SECOND_LOWEST_GRADES = []
GRADES = set()

for _ in range(int(input())):
    name = input()
    grade = float(input())
    STUDENTS.append([name, grade])
    GRADES.add(grade)

SECOND_LOWEST = sorted(GRADES)[1]

for name, grade in STUDENTS:
    if grade == SECOND_LOWEST:
        SECOND_LOWEST_GRADES.append(name)

for name in sorted(SECOND_LOWEST_GRADES):
    print(name, end='\n')
```

```python
# finding_the_percentage-1

if __name__ == '__main__':
    
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()

    def find_percentage(name):
        return f"{sum(student_marks[name]) / len(student_marks[name]):.2f}"

    print(find_percentage(query_name))
```

```python
# finding_the_percentage_2

from functools import reduce

if __name__ == '__main__':
    
    n = int(input())
    student_marks = {}
    for _ in range(n):
        name, *line = input().split()
        scores = list(map(float, line))
        student_marks[name] = scores
    query_name = input()

    marks = student_marks[query_name]
    sum_marks = reduce(lambda x, y: x + y, marks, 0)
    avg_marks = sum_marks/len(marks)
    print("{:.2f}".format(avg_marks))
```

```python
# finding_the_percentage_3

N = int(input())
STUDENT_MARKS = {}

for line in range(N):
    info = input().split(" ")
    grades = list(map(float, info[1:]))
    STUDENT_MARKS[info[0]] = sum(grades) / float(len(grades))

print("%.2f" % round(STUDENT_MARKS[input()], 2))
```

```python
# lists_1

if __name__ == '__main__':
    
    N = int(input())
    the_list = list()

    for _ in range(N):
        query = input().split()
        if query[0] == "print":
            print(the_list)
        elif query[0] == "insert":
            the_list.insert(int(query[1]), int(query[2]))
        elif query[0] == "remove":
            the_list.remove(int(query[1]))
        elif query[0] == "append":
            the_list.append(int(query[1]))
        elif query[0] == "sort":
            the_list = sorted(the_list)
        elif query[0] == "pop":
            the_list.pop()
        elif query[0] == "reverse":
            the_list.reverse()
```

```python
# lists_2

N = int(input())

ARRAY = []

while N != 0:
    A = input().split()

    if len(A) == 3:
        B = int(A[1])
        C = int(A[2])
    elif len(A) == 2:
        B = int(A[1])

    if A[0] == "insert":
        ARRAY.insert(B, C)
    elif A[0] == "print":
        print(ARRAY)
    elif A[0] == "remove":
        ARRAY.remove(B)
    elif A[0] == "append":
        ARRAY.append(B)
    elif A[0] == "sort":
        ARRAY.sort()
    elif A[0] == "pop":
        ARRAY.pop()
    elif A[0] == "reverse":
        ARRAY.reverse()
    N -= 1
```

```python
# sets_1

def average(array):
    '''Calculates the average of the array'''
    s = set(arr)
    return float(sum(s)) / len(s)

if __name__ == '__main__':
    
    n = int(input())
    arr = list(map(int, input().split()))
    result = average(arr)
    print(result)
```
